```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('lena.tiff')
```

```python
img_r,img_g,img_b = img.split()
```

```python
img1 = img_r.resize((50,50))
```

```python
img2 = img_g.transpose(Image.FLIP_LEFT_RIGHT).transpose(Image.ROTATE_270)
```

```python
img3 = img_b.crop((0,0,150,150))
```

```python
img4 = Image.merge("RGB",[img_r,img_g,img_b])
```

```python
plt.figure(figsize=(10,10))
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

plt.subplot(221)
plt.axis("off")
plt.imshow(img1)
plt.title("R-缩放",fontsize=14)

plt.subplot(222)
plt.imshow(img2)
plt.title("G-镜像+旋转",fontsize=14)

plt.subplot(223)
plt.axis("off")
plt.imshow(img3)
plt.title("B-裁剪",fontsize=14)

plt.subplot(224)
plt.axis("off")
plt.imshow(img4)
plt.title("RGB",fontsize=14)

plt.tight_layout()
plt.suptitle("图像基本操作",fontsize=20,color="blue",x=0.5,y=1.02)
```

```python
img4.save("lena.png")
```
